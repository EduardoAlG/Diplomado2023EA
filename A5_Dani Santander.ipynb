```python
import os
import pandas as pd

def import_data_sbs_banks(start_month, start_year, end_month=None, end_year=None):
    data_path = "_data/sbs/B_RawData/bancos"
    result_dict = {}

    if end_month is None or end_year is None:
        end_month = start_month
        end_year = start_year

    for year in range(start_year, end_year + 1):
        result_dict[year] = {}

        start_month_iter = start_month if year == start_year else 1
        end_month_iter = end_month if year == end_year else 12

        for month in range(start_month_iter, end_month_iter + 1):
            file_name = f"table_clean_{month}{month}{year}.xlsx"
            file_path = os.path.join(data_path, file_name)

            if os.path.exists(file_path):
                df = pd.read_excel(file_path)  # Asumiendo que los archivos son Excel
                result_dict[year][month] = df
            else:
                result_dict[year][month] = None

    return result_dict
```
